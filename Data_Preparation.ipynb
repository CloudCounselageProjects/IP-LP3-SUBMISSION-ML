## Creating a Dataframe based on the above dataset

```python
df_i = pd.read_csv('Amazon - Movies and TV Ratings.csv')

colunm_name = ['user','movie','rating']
df_f =pd.DataFrame(columns=colunm_name)
            
def add_data_to_df_f():

    str_movie = 'Movie'
    for i in range(4848):
        user_id=df_i['user_id'][i]
        for j in range(206):
            str_movie_no = str(j+1)
            
            if(df_i[str_movie+str_movie_no][i]!=0):
                df_f.loc[len(df_f.index)]=[user_id,j+1,df_i[str_movie+str_movie_no][i]]
        

add_data_to_df_f()
```

## Saving the new DataFrame as a .csv file
