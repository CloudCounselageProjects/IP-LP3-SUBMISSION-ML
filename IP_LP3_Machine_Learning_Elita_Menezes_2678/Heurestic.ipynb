```
img_path = 'img/'
```

```
import cv2
import numpy
```

```
img = cv2.imread(img_path + 'blue.jpg')
avg_color_per_row = numpy.average(img, axis=0)
avg_color = numpy.average(avg_color_per_row, axis=0)
print(avg_color)
```

```
color = avg_color.max()
print(color)
if color == avg_color[0]:
  print('Color : Blue')
elif color == avg_color[1]:
  print('Color : Green')
else:
  print('Color : Red')
```
