```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
dataset=pd.read_csv('Amazon - Movies and TV Ratings.csv')
dataset.head()
X=dataset.iloc[:,1:].values
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
#Taking Care of NaN values
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X[:,:])
X[:,:]=imputer.transform(X[:,:])
```

```python
print(X)
```

```python
df=dataset.melt(id_vars=["user_id"],var_name="MovieID",value_name="Ratings")
df
```

```python
b=df.iloc[:,1:2].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
b=np.array(ct.fit_transform(b))
```

```python
from sklearn.model_selection import train_test_split
b_train,b_test,X_train,X_test=train_test_split(b,X,test_size=0.2,random_state=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=10,random_state=0)
regressor.fit(b,X)
```
