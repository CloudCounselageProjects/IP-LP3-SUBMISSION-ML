```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
user_ratings = pd.read_csv('/content/sample_data/Amazon - Movies and TV Ratings.csv')
user_ratings
```

```python
#drop all the movies which have less than 10 user who rated it
user_ratings = user_ratings.fillna(0)
user_ratings
```

```python
train_data, test_data = train_test_split(user_ratings, test_size=0.2)
```

```python
test_data
```

```python
train_data
```

```python
item_similarity_dataframe = train_data.corr(method='pearson')
item_similarity_dataframe.head(50)
```

```python
def get_similar_movies(movie_name,train_data):
  similarity_score = item_similarity_dataframe[movie_name]*(train_data)
  similarity_score = similarity_score.sort_values(ascending = False)
  return similarity_score
```

```python
movies = [('Movie5',5)]
similar_movies = pd.DataFrame()

for movie,ratings in movies:
  similar_movies = similar_movies.append(get_similar_movies(movie,ratings),ignore_index=True)

similar_movies
similar_movies.sum().sort_values(ascending=False)
```
