# Movie Recomendation for users who have watched Movie1

## Required Libraries

```python
import pandas as pd
```

## Importing of the DataSet

```python
df = pd.read_csv('Amazon - Movies and TV Ratings.csv', index_col=0)
df.fillna(0, inplace=True)
df
```

## Standerdzation of the DataSet

```python
def standardize_df(row):
    new_row = (row - row.mean())/(10-(-1))
    return new_row

df_std = df.apply(standardize_df).T
df_std
```

## Cresting the CoRelation Matix

```python
corr_Matrix = df.corr(method='pearson')
corr_Matrix
```

```python
def get_similar(movie_name,rating):
    similar_rating = corr_Matrix[movie_name]*(rating-2.5)
    similar_rating = similar_rating.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_rating
```

```python
user_movie1 = [("Movie1",5)]
similar_rating = pd.DataFrame()
for movie,rating in user_movie1:
    similar_rating = similar_rating.append(get_similar(movie,rating),ignore_index = True)

similar_rating.head(10)
```

## List of Movies recomemded to User who has watched Movie1

```python
similar_rating.sum().sort_values(ascending=False)
```
