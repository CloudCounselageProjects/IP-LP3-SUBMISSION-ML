```python
#IMPORT LIBRARIES
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#READ DATASET
df = pd.read_csv('Amazon - Movies and TV Ratings.csv')
df.head()
```

```python
#FILLING NULL VALUES
df = df.fillna(0)
df.head()
```

```python
#CHANGING USER INDEXES INTO USER VALUES AND REMOVING USER_ID COLUMN
df.insert(0, "Users", ["User" + " " + str(i) for i in range(1,4849)])
df = df.drop("user_id", axis=1)
```

```python
#SETTING USER NAMES AS INDEX
df = df.set_index("Users")
```

```python
#STANDARDIZING EACH ROW IN DATAFRAME 
def standardize(row):
    new = row - (row - row.mean()) / (row.max() - row.min())
    return new

df_std = df.apply(standardize)
```

```python
#APPLYING COSINE SIMILARITY ON STANDARDIZED DATAFRAME
item_similarity = cosine_similarity(df_std.T)
```

```python
#CONVERTING STANDARDIZED NUMPY ARRAY TO DATAFRAME
item_similarity_df = pd.DataFrame(item_similarity, index=df.columns, columns=df.columns)
item_similarity_df #to view the similarity dataframe
```

```python
#FUNCTION TO GET SIMILARITY OF EACH MOVIE FOR SPECIFIC USER RATING
def similarity(movie, user_rating):
    similarity_ratings = item_similarity_df[movie]*(user_rating - 2.5)
    similarity_ratings = similarity_ratings.sort_values(ascending=False)
    return similarity_ratings
```

```python
#FUNCTION TO GET MOVIE RECOMMENDATION FOR PARTICULAR USER FROM ORIGINAL DATASET
def suggest_movies(user):
    movie = []
    rating = []
    for i in df.columns:
        if df[i][user] > 0:
            movie.append(i)
            rating.append(int(df[i][user]))
        merged  = [(movie[j], rating[j]) for j in range(0,len(movie))]
     
    similar_movies = pd.DataFrame()
    for movie,rating in merged:
        similar_movies = similar_movies.append(similarity(movie,rating), ignore_index= True)
        
    similar_movies = similar_movies.sum().sort_values(ascending=False) #most similar movies will be at the top
    return similar_movies[:10] #slicing to return only top 10 recommended movies
```

```python
#TESTING RECOMMENDATION FOR USER 1 (USER_ID = A3R5OBKS7OM2IR)
testing = suggest_movies("User 1") 
print(testing)                    
```
